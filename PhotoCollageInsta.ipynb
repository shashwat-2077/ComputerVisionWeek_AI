```python
# reading images from dataset
center = cv2.imread('/content/drive/MyDrive/Code/CV Week/Week 1 Day 2/center.jpeg')
center = cv2.cvtColor(center,cv2.COLOR_BGR2RGB)
bottom_left = cv2.imread('/content/drive/MyDrive/Code/CV Week/Week 1 Day 2/bottom_left (1).jpg')
bottom_left = cv2.cvtColor(bottom_left,cv2.COLOR_BGR2RGB)
bottom_right = cv2.imread('/content/drive/MyDrive/Code/CV Week/Week 1 Day 2/bottom_right.jpg')
bottom_right = cv2.cvtColor(bottom_right,cv2.COLOR_BGR2RGB)
top_left = cv2.imread('/content/drive/MyDrive/Code/CV Week/Week 1 Day 2/top_left.jpg')
top_left = cv2.cvtColor(top_left,cv2.COLOR_BGR2RGB)
top_right = cv2.imread('/content/drive/MyDrive/Code/CV Week/Week 1 Day 2/top_right.jpg')
top_right = cv2.cvtColor(top_right,cv2.COLOR_BGR2RGB)
```

```python
def create_blank_black_canvas(width, height):
    image = np.zeros((height, width, 3), np.uint8)
    return image
```

```python
canvas = create_blank_black_canvas(430, 430)
canvas_for_center = create_blank_black_canvas(120, 120)
plt.imshow(canvas)
plt.show()
plt.imshow(canvas_for_center)
plt.show()
```

```python
canvas_for_center[10:110,10:110] = center
plt.imshow(canvas_for_center)
plt.show()
```
