```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("/content/drive/MyDrive/Code/CV Week/jugal.png")
```

kernel are basically masks that are very impoimportant we we want to detect edges in a image


```python
kernel = np.ones((5,5), np.uint8)
```

# Basic Operations

```python
plt.imshow(img)
```

```python
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(imgRGB)
```

we put cmap = gray to convert the whwole image to graysacale otherwise it comes in some diffedifferent colour


```python
imgGray = cv2.cvtColor(imgRGB, cv2.COLOR_BGR2GRAY)
plt.imshow(imgGray, cmap = 'gray', vmin = 0, vmax = 255)
```

blurirng of image , it takes 3 parameters, the image, the kernel values and the sigma value that is the intensiy of the blur


```python
imgBlur = cv2.GaussianBlur(imgRGB, (7,7), 0)
```

```python
plt.imshow(imgBlur)
```

finding out the edges
parameters are the image and the threshold values

```python
imgCanny = cv2.Canny(imgRGB, 100, 100)
```

```python
plt.imshow(imgCanny)
```

refining the edges


```python
imgDilation = cv2.dilate(imgCanny, kernel, iterations=1)
```

```python
plt.imshow(imgDilation)
```

removing the noise of the image
3 parameters - image, the kernel and the no of iterations


```python
imgEroded = cv2.erode(imgDilation, kernel, iterations=1)
```

```python
plt.imshow(imgEroded)
```
