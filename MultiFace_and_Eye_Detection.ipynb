MULTI FACE DETECTION

```python
import cv2
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
image = cv2.imread("/content/drive/MyDrive/Code/CV Week/Week 2 Day 2/Team-India-Kit-2015_16_9_hd_1600.jpg");
plt.imshow(image);
```

```python
image_RGB=cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image_RGB)
```

```python
image_gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image_gray, cmap='gray')
```

```python
haar_cascade_face = cv2.CascadeClassifier('/content/drive/MyDrive/Code/CV Week/Week 2 Day 2/haarcascade_frontalface_default.xml')
```

```python

faces_rects = haar_cascade_face.detectMultiScale(image_gray, scaleFactor = 1.2, minNeighbors = 6);
print(faces_rects)
print('Faces found: ', len(faces_rects))
```

```python
for (x,y,w,h) in faces_rects:
     cv2.rectangle(image_RGB, (x, y), (x+w, y+h), (0, 0, 255), 20)
```

```python
plt.imshow(image_RGB);
```


EYE DETECTION

```python
eye_cascade = cv2.CascadeClassifier('/content/drive/MyDrive/Code/CV Week/Week 2 Day 2/haarcascade_eye_tree_eyeglasses.xml')
```

```python
eyes = eye_cascade.detectMultiScale(image_RGB)
for (ex, ey ,ew, eh) in eyes:
    cv2.rectangle(image_RGB, (ex,ey), (ex+ew, ey+eh), (0,0,255), 5)
```

```python
plt.imshow(image_RGB)
```
