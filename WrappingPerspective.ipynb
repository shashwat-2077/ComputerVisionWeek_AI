```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
imgBGR = cv2.imread("/content/drive/MyDrive/Code/CV Week/2/Screenshot 2021-06-23 200225.png")
```

```python
plt.imshow(imgBGR)
```

```python
img = cv2.cvtColor(imgBGR, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
width, height = 250, 350 # initializing the Standard playing cards size 
```

```python
points1 = np.float32([[223, 19], [386, 24], [389, 281], [216, 279]]) # Going clockwise/ creating a numpy array in datatype float 32. just open the image in mspaint and hover the cursor on the endpoints and just note the pixel values
```

```python
points2 = np.float32([[0, 0], [width, 0], [width, height], [0, height]]) #its the destination points which are used to map the final points to get a top down image based on the height and the width
```

```python
matrix = cv2.getPerspectiveTransform(points1, points2) # obtain the transformation of the image
```

```python
imgOutput = cv2.warpPerspective(img, matrix, (width, height)) # wrapping the image
```

```python
plt.imshow(img) 
```

```python
plt.imshow(imgOutput)
```
